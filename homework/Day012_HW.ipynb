### ① 取出今天所有的發文

```python
import requests 
from bs4 import BeautifulSoup
url='https://www.ettoday.net/news/news-list.htm'
r=requests.get(url)
re=r.text
soup=BeautifulSoup(re,'html5lib')
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find('em').text,d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import pandas as pd
import numpy as np
all=[]
for d in soup.find(class_="part_list_2").find_all('h3'):
    all.append([d.find('em').text,d.find(class_="date").text,d.find_all('a')[-1].text])
gp=pd.DataFrame(all,columns=['category','date','context'])
gp.sort_values(by='category')
```

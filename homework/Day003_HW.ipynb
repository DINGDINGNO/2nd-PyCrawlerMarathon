### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
#File I/O 將xml整個讀出來，而xmltodict則是以Dict的格式跑出來
import urllib.request
import zipfile
res="http://opendata.cwb.gov.tw/govdownload?dataid=F-D0047-093&authorizationkey=rdec-key-123-45678-011121314"
urllib.request.urlretrieve(res,".\data\example.zip")
f=zipfile.ZipFile(".\data\example.zip")
f.extractall(".\data")
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

import xmltodict
with open("./Data/64_Week24_CH.xml",encoding="utf-8")as fd:
    doc=xmltodict.parse(fd.read())
locals=doc["cwbopendata"]["dataset"]["locations"]["location"]
for local in locals:
    print(local["locationName"])
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
for i in locals:
    print(i['locationName'],"<",end='')
    print("攝氏度：", i['weatherElement'][0]['time'][0]['elementValue']['value'],"> ",end='')
    print("time:",i['weatherElement'][0]['time'][0]['startTime'])
 
   
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
map=locals[0]
name=map['locationName']
var=map['weatherElement'][0]['time']
for i in var :
    print(name,"< 溫度:",i['elementValue']['value'],"開始時間:",i['startTime'],"結束時間:",i['endTime'],">")

```

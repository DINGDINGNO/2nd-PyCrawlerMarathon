### 比較一下範例檔案中的「r.text」與「json.loads(r.text)」讀出來的內容有什麼差異

```python
#request的資料會以dic方式呈現，
import requests 
r= requests.get("http://odata.wra.gov.tw/v4/RealtimeWaterLevel")

r.text 
```

```python
import json

json.loads(r.text)
```
